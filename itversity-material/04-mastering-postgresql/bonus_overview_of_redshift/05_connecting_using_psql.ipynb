## Connecting using psql

Let us understand how we can connect to databases created in Redshift server using Postgres CLI `psql`.

* By default AWS will not let us to connect from out side systems such as our PCs.
* Here are the steps involved to connect to databases created in our AWS Redshift cluster.
  * Get the endpoint (from General Information of the Redshift cluster). We will use it later.
  * Make sure required databases and users are created as demonstrated in the previous topic. It is not advisable to use **master user** credentials to connect to databases in Redshift.
  * If you are using your system, you need to Postgres binaries installed. Redshift is a Postgres flavor.
  * Only when we have Postgres binaries installed we will be able to use `psql`.
  * Now you need to open up port used while creating the cluster by going to security group (AWS Firewall). In our case we have used 5439 which is default port number.
    * You can access security group under **properties** -> **Network and Security**.
    * You can open the port 5439 with your PC or Server IP (safest) or 0.0.0.0/0.
    * If you have created your cluster by using defaults for additional configurations, then the cluster is not publicly accessible.
    * Either you need to connect to cluster with in VPC or you need to make it publicly accessible.
    * In case you have created your cluster with default configurations, you can follow "Steps to make Redshift cluster publicly accessible"
  * Now you can connect to the cluster using `psql` with the following information.
    * Hostname: DNS from endpoint - **redshift-cluster-1.ckxblouy7rzo.us-east-1.redshift.amazonaws.com**
    * Port Number: 5439
    * Database: retail_db
    * Username: retail_user
    * Password: **Your Password** (in my case it is Retail_p@ssw0rd)
  * Here is the sample command to connect to Redshift database using `psql`. It will prompt for the password and make sure to enter the right password.
  
```shell
psql -h redshift-cluster-1.ckxblouy7rzo.us-east-1.redshift.amazonaws.com \
  -p 5439 \
  -d retail_db \
  -U retail_user \
  -W
```

* You can setup tables and load the data following these steps.
  * Clone GitHub repository to the desired location.
  * Run the below script to create tables. Make sure to use correct location of the script as per your environment.

```shell
psql -h redshift-cluster-1.ckxblouy7rzo.us-east-1.redshift.amazonaws.com \
  -p 5439 \
  -d retail_db \
  -U retail_user \
  -W \
-f ~/Research/data/retail_db/create_db_tables_pg.sql
```

  * Run the below script to load data in the tables. Make sure to use correct location of the script as per your environment.

```shell
psql -h redshift-cluster-1.ckxblouy7rzo.us-east-1.redshift.amazonaws.com \
  -p 5439 \
  -d retail_db \
  -U retail_user \
  -W \
-f ~/Research/data/retail_db/load_db_tables_pg.sql
```

* Connect to the database using `psql`.

```shell
psql -h redshift-cluster-1.ckxblouy7rzo.us-east-1.redshift.amazonaws.com \
  -p 5439 \
  -d retail_db \
  -U retail_user \
  -W
```
